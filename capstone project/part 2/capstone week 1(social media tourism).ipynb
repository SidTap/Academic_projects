```python
df.following_company_page = df.following_company_page.replace('1','Yes')
df.following_company_page = df.following_company_page.replace('0','No')
```

```python
df.following_company_page.unique()  
```

```python
### preferred_location_type 
```

```python
df.preferred_location_type.unique()  
```

```python
df.preferred_location_type = df.preferred_location_type.replace('Tour and Travel','Tour  Travel')
df.preferred_location_type = df.preferred_location_type.replace('OTT','Other')
df.preferred_location_type = df.preferred_location_type.replace('Movie','Other')
```

```python
df.preferred_location_type.value_counts() 
```

```python
### member_in_family  
```

```python
df.member_in_family.value_counts()  
```

```python
df.member_in_family = df.member_in_family.replace('Three','3')
df.member_in_family = df.member_in_family.replace('10','5')
```

```python
df.member_in_family.value_counts()  
```

```python
df.preferred_device.value_counts()  
```

```python
df.preferred_device = df.preferred_device.replace('Android OS','Android')
df.preferred_device = df.preferred_device.replace('ANDROID','Android')
df.preferred_device = df.preferred_device.replace('Others','Other')
```
