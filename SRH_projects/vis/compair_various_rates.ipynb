```python
import json

# Load the JSON data
def load_json_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Extract the current prices of cryptocurrencies
def extract_prices(data):
    prices = {}
    for coin in data:
        coin_id = coin.get('id')
        usd_price = coin.get('current_price', None)
        if coin_id and usd_price is not None and usd_price > 0:
            prices[coin_id] = usd_price
    return prices

# Perform pairwise comparisons
def compare_all_currencies(prices):
    comparisons = {}
    for coin1 in prices:
        comparisons[coin1] = {}
        for coin2 in prices:
            if coin1 != coin2:
                ratio = prices[coin1] / prices[coin2]
                comparisons[coin1][coin2] = ratio
    return comparisons

# Main function
def main():
    file_path = 'C:/Users/siddd/OneDrive/Desktop/srh/data visualisation 3/final project/CoinGecko/coins_market_data.json'  # Replace with your file path
    data = load_json_data(file_path)
    prices = extract_prices(data)
    
    # Comparing every currency with every other currency
    all_comparisons = compare_all_currencies(prices)

    # Print the comparisons
    for coin1, comparisons in all_comparisons.items():
        for coin2, ratio in comparisons.items():
            print(f"1 {coin1.upper()} is equivalent to {ratio:.6f} {coin2.upper()}")

if __name__ == "__main__":
    main()


```
